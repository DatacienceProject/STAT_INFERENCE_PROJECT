```python
def clean_string_data(price_str):
    if pd.isna(price_str):
        return np.nan
        
    price_str = str(price_str).strip()

    clean_price_str = price_str.replace('$','').replace(',','').strip()

    parts = []
    if '-' in clean_price_str:
        parts = clean_price_str.split('-')
    elif '–' in clean_price_str:
        parts = clean_price_str.split('–')
    else:
        parts = [clean_price_str]

    numbers = []
    for i in parts:
        try:
            numbers.append(float(i.strip()))
        except ValueError:
            continue
            
    if len(numbers) == 2:
        return (numbers[0]+numbers[1])/2
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return np.nan
```

```python
# Correct way to test with a list of values
test_prices_list = ['$12,000', '$14,000-$20,000', '$9,000', '$3,000–$5,000']

# Convert the list to a pandas Series
test_prices_series = pd.Series(test_prices_list)
```
